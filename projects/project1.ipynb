## Load Libraries and Data


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

path = "..\\data"

# Load both sheets
df_2010_2011 = pd.read_excel(os.path.join(path,'online_retail_II.xlsx'), sheet_name='Year 2010-2011')
df_2009_2010 = pd.read_excel(os.path.join(path,'online_retail_II.xlsx'), sheet_name='Year 2009-2010')

# Display the first few rows of each dataset
print("Year 2010-2011:")
print(df_2010_2011.head())
print("\nYear 2009-2010:")
print(df_2009_2010.head())
```

```python
# Combine the datasets
df_combined = pd.concat([df_2009_2010, df_2010_2011], axis=False)

# Display the first few rows of the combined dataset
print("Combined Data:")
print(df_combined.head())
```

```python
# General information about the combined dataset
df_combined.info()

# Statistical description of the combined dataset
df_combined.describe()

# Clean data (optional)
df_combined = df_combined.dropna(subset=['Customer ID'])  # Drop rows without Customer ID
df_combined['InvoiceDate'] = pd.to_datetime(df_combined['InvoiceDate'])  # Convert dates to datetime format
```

```python
# Visualize the distribution of important variables
sns.countplot(x='Country', data=df_combined)
plt.xticks(rotation=90)
plt.show()
```
